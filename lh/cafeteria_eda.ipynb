```python
import os
import pandas as pd
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
train_df = pd.read_csv('./data/train.csv')
```

```python
train_df.info()
```

```python
train_df['일자'] = pd.to_datetime(train_df['일자'])
```

```python
train_df
```

```python
# 0: 월, 1: 화, 2: 수, 3: 목, 4: 금
train_df['요일'] = train_df['일자'].dt.dayofweek
```

```python
train_df['조식_메뉴1'] = train_df['조식메뉴'].map(lambda x: x.split()[0])
train_df['조식_메뉴2'] = train_df['조식메뉴'].map(lambda x: x.split()[1])
train_df['조식_메뉴3'] = train_df['조식메뉴'].map(lambda x: x.split()[2])
train_df['조식_메뉴4'] = train_df['조식메뉴'].map(lambda x: x.split()[3])
train_df['조식_메뉴5'] = train_df['조식메뉴'].map(lambda x: x.split()[4])
train_df['조식_메뉴6'] = train_df['조식메뉴'].map(lambda x: x.split()[5])
train_df['조식_메뉴7'] = train_df['조식메뉴'].map(lambda x: x.split()[6])
train_df['조식_메뉴8'] = train_df['조식메뉴'].map(lambda x: x.split()[7])
```

```python
train_df['중식_메뉴1'] = train_df['중식메뉴'].map(lambda x: x.split()[0])
train_df['중식_메뉴2'] = train_df['중식메뉴'].map(lambda x: x.split()[1])
train_df['중식_메뉴3'] = train_df['중식메뉴'].map(lambda x: x.split()[2])
train_df['중식_메뉴4'] = train_df['중식메뉴'].map(lambda x: x.split()[3])
train_df['중식_메뉴5'] = train_df['중식메뉴'].map(lambda x: x.split()[4])
train_df['중식_메뉴6'] = train_df['중식메뉴'].map(lambda x: x.split()[5])
train_df['중식_메뉴7'] = train_df['중식메뉴'].map(lambda x: x.split()[6])
```
